## Handling Null Values

```python
col = list(insurance.columns)
col
```

```python
for coln in col:
    if coln == 'children':
        insurance[coln] = insurance[coln].fillna(insurance[coln].mode()[0])
    elif insurance[coln].dtype == 'object':
        insurance[coln] = insurance[coln].fillna(insurance[coln].mode()[0])
    else:
        insurance[coln] = insurance[coln].fillna(insurance[coln].mean())
```

```python
insurance.isnull().sum()
```

## Handling The Outliers

```python
for coln in col:
    if(insurance[coln].dtype == 'float64'):
        plt.boxplot(insurance[coln])
        plt.xlabel(coln)
        plt.show()
```

```python
columns = ['bmi','past_consultations','Hospital_expenditure','Anual_Salary']
for col in columns:
  Q1 = insurance[col].quantile(0.25)
  Q3 = insurance[col].quantile(0.75)
  IQR = Q3 - Q1
  lowerFence = Q1-(1.5*IQR)
  upperFence = Q3 +(1.5*IQR)
  insurance = insurance[(insurance[col] >= lowerFence) & (insurance[col]<=upperFence)]
  plt.boxplot(insurance[col])
  plt.xlabel(coln)
  plt.show()
```

## Encoding of the dataset

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_colms = insurance.select_dtypes(include = ['object']).columns
cat_colms
```

```python
la = LabelEncoder()
```

## Preparing The Model

```python
from sklearn.model_selection import train_test_split
```

```python
x = insurance.drop('charges' , axis = 1)
y = insurance['charges']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.8,random_state=0)
```
